```
from pyspark.sql.functions import *
```

```
path = "/FileStore/tables/weatherstats_toronto_daily.csv"
table = spark.read.format("csv").option("header", True).load(path)

table.printSchema()
```

```
table.createOrReplaceTempView("temp")
df = spark.sql("SELECT date Date, rain Rain, snow Snow, snow_on_ground SnowOnGround, precipitation Precipitation, avg_visibility AvgVisibility, min_visibility MinVisibility, avg_wind_speed AvgWind, max_wind_speed MaxWind from temp")

df = df.withColumn("Date", to_date(col("Date"), "yyyy-MM-dd")) \
        .withColumn("Year", year(col("Date"))) \
        .withColumn("Month", month(col("Date"))) \
        .withColumn("Rain", col("Rain").cast("decimal(10,1)")) \
        .withColumn("Snow", col("Snow").cast("decimal(10,1)")) \
        .withColumn("SnowOnGround", col("SnowOnGround").cast("int")) \
        .withColumn("Precipitation", col("Precipitation").cast("decimal(10,1)")) \
        .withColumn("AvgVisibility", col("AvgVisibility").cast("decimal(10,1)")) \
        .withColumn("MinVisibility", col("MinVisibility").cast("int")) \
        .withColumn("AvgWind", col("AvgWind").cast("decimal(10,1)")) \
        .withColumn("MaxWind", col("MaxWind").cast("int"))

display(df)
```
